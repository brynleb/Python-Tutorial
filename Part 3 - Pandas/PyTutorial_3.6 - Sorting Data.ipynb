In this tuturial we will learn how to sort data in our DataFrames.

```python
people = {
    'first' : ['Corey' , 'Jane' , 'John', 'Tony'],
    'last' : ['Schafer' , 'Doe' , 'Doe', 'Smith'],
    'email' : ['CoreyMShafer@email.com' , 'JaneDoe@email.com' , 'JohnDoe@email.com', 'TonySmith@email.com'],
}

import pandas as pd 
df = pd.DataFrame(people)
df
```

Let's try sorting the df by last name 

```python
df.sort_values(by='last',ascending=False)
```

Since we have a last name duplicated, let's sort by last names, then first names.

```python
df.sort_values(by=['last', 'first'],ascending=False)
```

It is also possible to make the second variable ascending while the first is descending, and vice versa. 

```python
df.sort_values(by=['last', 'first'],ascending=[False, True])
```

We can set any of these search perametres to become defualt for the DataFrame, untill changed. 

```python
df.sort_values(by=['last', 'first'],ascending=[False, True], inplace=True)
df
```

We can also sort by index, returing the DataFrame to it's defualt state

```python
df.sort_index()
```
